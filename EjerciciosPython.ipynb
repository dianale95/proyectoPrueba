```python
import random

# Usuario elige un número
numero_persona = int(input("Dime un número: "))

# Aseguramos que el número esté dentro de un rango válido
while numero_persona < 1 or numero_persona > 3:
    print("El número debe estar entre 1 y 3")
    numero_persona = int(input("Dime un número: "))

# Máquina intenta adivinar si el número es par o impar
adivinanza_maquina = random.choice(["par", "impar"])

# Comprobamos si el número de la persona es par o impar
if numero_persona % 2 == 0:
    resultado_persona = "par"
else:
    resultado_persona = "impar"

print(f"El número de la persona es: {numero_persona} ({resultado_persona})")
print(f"La máquina adivinó que el número es: {adivinanza_maquina}")

# Si la máquina acierta
if adivinanza_maquina == resultado_persona:
    print("¡La máquina ha acertado! Tú has perdido.")
else:
    print("¡La máquina ha fallado! ¡Ganaste!")
```

```python
# Definimos el alfabeto personalizado
alfabeto = "abcdefghijklmnopqrstuvwxyz"

# Función para encriptar con ROT13
def rot13(mensaje):
    mensaje_encriptado = ""

    for letra in mensaje:
        if letra in alfabeto:  # Verificamos si la letra está en el alfabeto
            # Encontramos la posición de la letra en el alfabeto
            indice = alfabeto.index(letra)
            # Aplicamos ROT13, sumando 13 al índice y tomando el módulo con la longitud del alfabeto
            nueva_pos = (indice + 13) % len(alfabeto)
            mensaje_encriptado += alfabeto[nueva_pos]
        else:
            # Si la letra no está en el alfabeto (como un espacio o puntuación), la dejamos tal cual
            mensaje_encriptado += letra

    return mensaje_encriptado

# Ingresar mensaje a encriptar
mensaje = input("Dime un mensaje: ")

# Mostrar el mensaje encriptado con ROT13
mensaje_encriptado = rot13(mensaje)
print("Mensaje encriptado con ROT13:", mensaje_encriptado)
```

```python
import hashlib

# Implementar md5

mensaje_original = input("Dime un mensaje: ")
# Para usar MD5, debemos usar el módulo hashlib
mensaje_bytes = mensaje_original.encode('utf-8') # Convertir el mensaje a bytes (Español)
mensaje_cifrado = hashlib.md5(mensaje_bytes).hexdigest() # Obtener el hash MD5

print("El mensaje original es: ", mensaje_original)
print("El mensaje cifrado es: ", mensaje_cifrado)
```

```python
ruta_compras = "/content/compras_semana.csv"

datos_compras = pd.read_csv(ruta_compras)
# Aumenta columna total
datos_compras['total'] = datos_compras['precio'] * datos_compras['cantidad']

#Suma columna total
suma_compra = datos_compras['total'].sum().round(2)
print(f"Total de compra: {suma_compra}")

#Promedio
promedio_compra = datos_compras['total'].mean()
print(f"Promedio de compra: {promedio_compra}")

#Mediana
mediana_compra = datos_compras['total'].median()
print(f"Mediana de compra: {mediana_compra}")

#Compra mínima
compra_minima = datos_compras['total'].min()
print(f"Compra mínima: {compra_minima}")

#Desviación estándar
desviacion_compra = datos_compras['total'].std()
print(f"Desviación estándar de compra: {desviacion_compra}")

#Registro conteo
conteo_compra = datos_compras['total'].count()
print(f"Conteo de compra: {conteo_compra}")

#Hace el resumen de todas las medidas
datos_compras['total'].describe()

#datos_compras.head(10)
```
